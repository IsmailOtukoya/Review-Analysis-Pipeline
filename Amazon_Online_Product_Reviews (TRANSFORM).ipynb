```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
df = pd.read_csv(r"C:\Users\ISMAIL\OneDrive\Documents\data science projects\cleaned_dataset_AOPR.csv")
```

```python
df.head()
```

```python
# Drop rows with NaN values in the 'reviews.text' column
df.dropna(subset=['reviews.text'], inplace=True)


```

```python
# Extract the review text column
reviews = df['reviews.text'].tolist()
vectorizer = CountVectorizer(max_features=3000, min_df=5, stop_words='english', ngram_range=(1, 2), binary=True, lowercase=True)
dtm = vectorizer.fit_transform(reviews)
```

```python
feature_names = vectorizer.get_feature_names_out()
dtm_df = pd.DataFrame(dtm.toarray(), columns=feature_names)
```

```python
df_transformed = pd.concat([df, dtm_df], axis=1)

df_transformed.head()
```

```python
df_transformed.to_csv('transformed_dataset1.csv', index=False)  
```
