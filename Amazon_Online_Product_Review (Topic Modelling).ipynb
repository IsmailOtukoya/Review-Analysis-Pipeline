# Topic Modelling

##### Importing Libraries:

##### pandas (imported as pd): Used for data manipulation and analysis.
##### LatentDirichletAllocation from sklearn.decomposition: Used for Latent Dirichlet Allocation (LDA) topic modeling.
##### NMF from sklearn.decomposition: Used for Non-Negative Matrix Factorization (NMF) topic modeling.
##### matplotlib.pyplot (imported as plt): Used for data visualization.
##### WordCloud from wordcloud: Used for generating word clouds.

##### Loading the Dataset:

##### The read_csv() function from pandas is used to read the CSV file into a pandas DataFrame named df. The file path is provided as an argument to the function.
##### Preparing the Data:

##### The iloc indexer is used to select the columns of interest from the DataFrame, starting from the 9th column. This assumes that these columns represent the document-term matrix (DTM) or a similar representation.
##### The dropna() method is used to remove any rows with missing values from the DTM.
##### Topic Modeling with LDA:

##### An instance of the LatentDirichletAllocation class is created, specifying the number of topics (10) and a random state for reproducibility.
##### The fit_transform() method is called on the LDA model, passing the DTM as input. This step applies the LDA algorithm to the DTM and obtains the topic distributions for each document.
##### A loop iterates over each topic, accessing the topic-word distributions from the components_ attribute of the LDA model. The top words for each topic are extracted using the argsort() method, and the corresponding feature names are retrieved from the DTM columns. The top words are then printed for each topic.
##### Topic Modeling with NMF:

##### An instance of the NMF class is created, specifying the number of topics (10) and a random state for reproducibility.
##### The fit() method is called on the NMF model, passing the DTM as input. This step applies the NMF algorithm to the DTM and learns the topic-term distributions.
###### A loop iterates over each topic, accessing the topic-term distributions from the components_ attribute of the NMF model. The top words for each topic are extracted using the argsort() method, and the corresponding feature names are retrieved from the DTM columns. The top words are then printed for each topic.
##### Generating Word Clouds:

##### topic_keywords is a list of lists where each inner list contains the most important words for a specific topic. These words are manually defined based on the top words obtained from the LDA and NMF models.
##### A loop iterates over each topic and generates a word cloud using the WordCloud class. The word cloud is displayed using matplotlib.pyplot.
##### The code demonstrates how to perform topic modeling using LDA and NMF algorithms and visualize the results using word clouds. It can be useful for exploring and understanding the main topics present in a given dataset.






```python
import pandas as pd
from sklearn.decomposition import LatentDirichletAllocation
```

```python
df = pd.read_csv(r"C:\Users\ISMAIL\OneDrive\Documents\data science projects\transformed_dataset1.csv")
```

```python
dtm = df.iloc[:, 9:]  
dtm = dtm.dropna()
```

```python
lda_model = LatentDirichletAllocation(n_components=10, random_state=42)  
lda_output = lda_model.fit_transform(dtm)
```

```python
# Get the list of feature names (words)
feature_names = dtm.columns

# Define the number of top words to display for each topic
top_words = 10

# Iterate over each topic
for topic_idx, topic in enumerate(lda_model.components_):
    # Get the top words with highest probabilities for the current topic
    top_words_indices = topic.argsort()[:-top_words - 1:-1]
    top_words_list = [feature_names[i] for i in top_words_indices]
    
    # Print the topic and its top words
    print(f"Topic #{topic_idx + 1}:")
    print(", ".join(top_words_list))
    print()
```

```python
from sklearn.decomposition import NMF
```

```python
nmf_model = NMF(n_components = 10, random_state = 42)
```

```python
nmf_model.fit(dtm)
```

```python


# Iterate over each topic
for topic_idx, topic in enumerate(nmf_model.components_):
    # Get the top words with highest probabilities for the current topic
    top_words_indices = topic.argsort()[:-top_words - 1:-1]
    top_words_list = [feature_names[i] for i in top_words_indices]
    
    # Print the topic and its top words
    print(f"Topic #{topic_idx + 1}:")
    print(", ".join(top_words_list))
    print()
```
